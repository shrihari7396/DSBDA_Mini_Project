# DATA CLEANING

```python
# remove state india
df = df[df['State'] != 'India']
```

```python
#remove extra empty column
df.columns = df.columns.str.strip()
```

```python
#Find the null values
print(df.isnull().sum())
```

```python
#Fill the null values with 0
df.fillna(0, inplace=True)
```

```python
# handle dublicate rows
df.drop_duplicates(inplace=True)
```

## B) TOATAL VACCINATED

```python
 # B.1 Number of Persons State-wise Vaccinated (First Dose)
first_dose = df.groupby('State')['First Dose Administered'].max()
print(first_dose)

df_latest = df.sort_values('Updated On').groupby('State').tail(1)

#  plot
plt.figure()
plt.bar(df_latest['State'], df_latest['First Dose Administered'])
plt.xticks(rotation=90)
plt.title("First Dose State-wise")
plt.xlabel("State")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
# B.2  Number of Persons State-wise Vaccinated (Second Dose)
second_dose = df.groupby('State')['Second Dose Administered'].max()
print(second_dose)



df_latest2 = df.sort_values('Updated On').groupby('State').tail(1)

# plot
plt.figure()
plt.bar(df_latest2['State'], df_latest2['Second Dose Administered'])
plt.xticks(rotation=90)
plt.title("Second Dose State-wise")
plt.xlabel("State")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```
